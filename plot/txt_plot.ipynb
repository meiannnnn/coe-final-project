```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
def txt_plot(txt_file_path, title):
    with open(txt_file_path, 'r', encoding='latin-1') as file:
        txt_content = file.readlines()

    txt_data = []
    for line in txt_content:
        if line.strip() and not line.startswith(('X', 'Displacement', 'Force')):
            values = line.split()
            if len(values) == 3:
                try:
                    displacement = float(values[1])
                    force = float(values[2])
                    txt_data.append([displacement, force])
                except ValueError:
                    continue
    
    txt_data = np.array(txt_data)

    txt_data_converted = txt_data.copy()
    txt_data_converted[:, 1] = txt_data[:, 1] / 1000

    plt.figure(figsize=(8, 6))
    plt.plot(txt_data_converted[:, 0], txt_data_converted[:, 1], label="Simulated data (converted to kN)", color='b')

    plt.title(title)
    plt.xlabel("Displacement (mm)")
    plt.ylabel("Force (kN)")
    plt.grid(True)
    plt.legend()
    plt.xlim(left=0)
    plt.ylim(bottom=0)

    plt.show()
```

```python
folder_path = 'C:/Users/meian/Desktop/Comp Eng Project/coe-final-project/coe-final-project/plot/Tmp233.15K_StrRt0.0001_TD'

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        txt_plot(file_path, f"Tmp233.15K_StrRt0.0001 {filename}")
```

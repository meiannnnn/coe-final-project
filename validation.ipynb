```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score
```

```python
file_path = 'extended_curve/Tmp473.15K_StrRt0.001_RD.csv'
data_exp = pd.read_csv(file_path)

model_file_path = 'Temperature_calibration/Final/6 parameter/True_stress_Combined_473.15K.csv'
data_model = pd.read_csv(model_file_path)

```

```python
data_exp_filtered = data_exp[data_exp["Strain"] <= 0.3]
# Filter the experimental data for every 3rd element to create a scatter plot
data_exp_scatter = data_exp_filtered.iloc[::5]

# Plot experimental data (scatter) and adjusted model data for comparison
plt.figure(figsize=(10, 6))

# Plot experimental data as scatter
plt.scatter(data_exp_scatter["Strain"], data_exp_scatter["Stress"], label="Experimental Data")

# Plot adjusted model data
plt.plot(data_model["PEEQ"], data_model["True stress"] / 0.5 * 0.4, label="Model Prediction")

# Add labels and legend
plt.title("Stress-Strain Comparison for Unified Model")
plt.xlabel("True Strain, -")
plt.ylabel("True Stress, MPa")
plt.legend()
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_model["PEEQ"], data_model["stress (temperature calibration)"])
plt.scatter(data_exp_scatter["Strain"], data_exp_scatter["Stress"], label="Experimental Data")

plt.title("Temperature Calibration data")
plt.xlabel("True Strain, -")
plt.ylabel("True Stress, MPa")
plt.legend()
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_model["PEEQ"], data_model["stress (strain rate calibration)"])
plt.scatter(data_exp_scatter["Strain"], data_exp_scatter["Stress"], label="Experimental Data")

plt.title("Strain Rate calibration data")
plt.xlabel("True Strain, -")
plt.ylabel("True Stress, MPa")
plt.legend()
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_model["PEEQ"], data_model["stress (temperature calibration)"])
plt.plot(data_model["PEEQ"], data_model["stress (strain rate calibration)"])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data_model["PEEQ"], (data_model["stress (temperature calibration)"] + data_model["stress (strain rate calibration)"]) / 2)
plt.scatter(data_exp_scatter["Strain"], data_exp_scatter["Stress"], label="Experimental Data")
plt.show()
```
